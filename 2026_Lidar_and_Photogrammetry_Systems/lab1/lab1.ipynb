```python
import laspy
import numpy as np
import open3d as o3d
```

```python
data_2012 = laspy.read("2012/golm_364806_fp.las")
```

```python
points = np.vstack((data_2012.x, data_2012.y, data_2012.z)).T
```

```python
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points)
o3d.visualization.draw_geometries([pcd])
```
